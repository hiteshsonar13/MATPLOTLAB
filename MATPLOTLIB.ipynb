```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
### Draw a line in a diagram from position (o,10) to position (5,150)
xpoint=np.array([1,2])
ypoint=np.array([5,10])

plt.plot(xpoint,ypoint)
plt.show()
```

```python
x_point=np.array([1,20])
y_point=np.array([10,200])
plt.plot(x_point,y_point)
plt.show()
```

```python
xpoint=np.array([1,6])
ypoint=np.array([4,5])
plt.plot(xpoint,ypoint,"+")
plt.show()
```

```python
ypoint=np.array([3,8,1,10])
plt.plot(ypoint,'*''r')
plt.show()
```

```python
ypoints=np.array([3,8,1,12])
xpoints=np.array([5,6,7,12])
plt.plot(xpoints,ypoints, marker='*', ms=10, color='k')
plt.show()
### ms (stand for marker size)in color 'k' stand for black color
```

```python
ypoints=np.array([3,8,1])

plt.plot(ypoints, marker='o', ms=10 , mec='c',color='pink')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])

plt.plot(ypoints, marker='o',ms=10, mfc='r', mec='k')
plt.show()
```

```python
### use both the mec and mfc arguments to colour of the entires marker
ypoints=np.array([3,8,1,10])

plt.plot(ypoints,marker='o',ms= 10,mec='y',mfc='k')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])

plt.plot(ypoints,linestyle='dashed',marker='o', mec='hotpink',ms=20,mfc='green',linewidth=1)
plt.show()
```

```python
xpoint=np.array([1,6])
ypoint=np.array([4,5])

plt.plot(xpoint,ypoint,"+")
plt.show()
```

```python
### you can use keyboard argument linewidth or the shorter lw to change the width of the line. 

### The value is floating numebr, in points
ypoints=np.array([3,8,1,10])
xpoints=np.array([2,4,6,8])
plt.plot(xpoints,ypoints,linewidth=1,linestyle="dotted",color="red",marker="*",ms=10,mec="k")
plt.show()
```

```python
x=np.array([80,84,90,95,100,105,110,115,120,134])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y,color='r',linewidth=1)

plt.xlabel("Average Pulse", fontsize=10,color="red")
plt.ylabel("Calorie Burnage",fontsize=10,color="green")
plt.title("SAMPLE",fontsize=30,color='blue')
plt.show()
```

```python
### you can use the axis parameter in the grid() function to specify  which grid lines to display 

### Legal values are:'x','y', and 'both'. Default value is 'both'
x=np.array([80,84,90,95,100,105,110,115,120,134])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x,y)

plt.grid(axis='x')

plt.show()
```

```python
### set line properties for the grid
### you can also set the line properties of the grid, like this: grid
        ### (color = 'color',linestyle = 'linwestyle', linewidth=number)
x=np.array([80,84,90,95,100,105,110,115,120,134])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x,y)

plt.grid(color='hotpink',linestyle='--', linewidth=1.5)

plt.show()
```
